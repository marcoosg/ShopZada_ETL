## Step 2: Create Fact Table

```python
# Merge dataframes to create the fact table
fact_table = pd.merge(df_order_data[['ORDER_ID', 'USER_ID', 'PRODUCT_ID', 'QUANTITY', 'STAFF_ID', 'ESTIMATED_ARRIVAL']],
                      df_user_dimension[['USER_ID']],
                      how='inner',
                      on='USER_ID')

fact_table = pd.merge(fact_table,
                      df_product_dimension[['PRODUCT_ID']],
                      how='inner',
                      on='PRODUCT_ID')

fact_table = pd.merge(fact_table,
                      df_staff_dimension[['STAFF_ID']],
                      how='inner',
                      left_on='ORDER_ID',
                      right_on='STAFF_ID')

fact_table = pd.merge(fact_table,
                      df_merchant_dimension[['MERCHANT_ID']],
                      how='inner',
                      left_on='ORDER_ID',
                      right_on='MERCHANT_ID')

fact_table = pd.merge(fact_table,
                      df_campaign_dimension[['CAMPAIGN_ID']],
                      how='inner',
                      left_on='ESTIMATED_ARRIVAL',  # Assuming CAMPAIGN_ID is related to ESTIMATED_ARRIVAL
                      right_on='CAMPAIGN_ID')

# Select the required columns for the fact table
fact_table = fact_table[['ORDER_ID', 'USER_ID', 'PRODUCT_ID', 'QUANTITY', 'STAFF_ID', 'MERCHANT_ID', 'CAMPAIGN_ID']]
```
